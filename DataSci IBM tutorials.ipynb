Q1: Which discipline and event has awarded the most gold medals?

```python
gold_df = medals_df[medals_df.Medal=='Gold']  #filter for gold countries
```

```python
#futher filter for events that won gold
by_event = gold_df.groupby(['Discipline', 'Event']) 
```

```python
by_event.head()  #to get a glimpse
```

```python
#make a tally of each gold event
golds_by_event = by_event.Medal.count() 
golds_by_event.sort(ascending=False)
golds_by_event.head(10)
```

```python
golds_by_event.max()
```

#Q2 Which country  has the most gold silver and bronze medals?

```python
medals_by_country = medals_df.groupby(['NOC','Medal']).size()
medals_by_country.head()
```

```python
#convert from series to a dataframe
medals_by_country_df = medals_by_country.unstack()
medals_by_country_df.head()
```

```python
#replace missing data with zero value for countries which dont medal in eveyr colour
medals_by_country_df.fillna(0,inplace=True)
```

```python
#use idxmax() that returns maximum count (country code) for each medal
medals_by_country_df.idxmax()
```

```python
%matplotlib inline 
```

```python
#sort by highest 
medals_by_country_df.sort('Gold', ascending=False, inplace=True ) 
#plot top 15 countries
medals_by_country_df[['Gold','Silver', 'Bronze',]][:15].plot(kind='bar',figsize=(12,10))
```

```python
#function returns countires taht obtained at least 5 gold medals
golds_by_event_df = golds_by_event.unstack()

medals_by_country_df['Five Plus']= medals_by_country_df['Gold'].apply(lambda x: x >= 5)
medals_by_country_df['Five Plus'].count
```

```python
medals_by_country_df['Five Plus'].value_counts() #boolean index
```

```python
#calculate percentage of 5plus gold medalers
under_five = medals_by_country_df['Five Plus'].value_counts()[0]#returns first value which all false
five_plus = medals_by_country_df['Five Plus'].value_counts()[1] #returns second value which all true
print(float(five_plus/(under_five+five_plus)*100), '%')
```
